```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

## Player Count

* Display the total number of players


```python
total_player_count = len(purchase_data["SN"].value_counts())
total_player_df = pd.DataFrame({
    "Total Players": [total_player_count]
})

total_player_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
purchasing_analysis_df = pd.DataFrame({
    "Number of Unique Items":[len(purchase_data["Item ID"].value_counts())],
    "Average Price": [purchase_data["Price"].mean()],
    "Number of Purchases":[purchase_data["Purchase ID"].count()],
    "Total Revenue": [purchase_data["Price"].sum()]
})

purchasing_analysis_df["Average Price"] = purchasing_analysis_df["Average Price"].map("${:.2f}".format)
purchasing_analysis_df["Total Revenue"] = purchasing_analysis_df["Total Revenue"].map("${:.2f}".format)
purchasing_analysis_df.round(2)
```

##Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
genddem_df = pd.DataFrame()
genddem_df["Total Players"] = purchase_data.groupby(["Gender"])["SN"].nunique()
genddem_df["Percentage of Players"] = (purchase_data.groupby(["Gender"])["SN"].nunique()/
                                       purchase_data.groupby(["Gender"])["SN"].nunique().sum())*100
genddem_df["Percentage of Players"] = genddem_df["Percentage of Players"].map("{:.2f}%".format)
genddem_df.sort_values(by=["Total Players"], ascending=False)
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purgen_df = pd.DataFrame()

purchase_count = purchase_data["Gender"].value_counts()
purgen_df["Purchase Count"] = purchase_count

avg_purchase = purchase_data.groupby(["Gender"])["Price"].mean().round(2).map("${:.2f}".format)
purgen_df["Average Purchase Price"] = avg_purchase

total_purchase = purchase_data.groupby(["Gender"])["Price"].sum().round(2).map("${:.2f}".format)
purgen_df["Total Purchase Value"] = total_purchase

per_person = purchase_data.groupby(["Gender"])["Price"].sum()/purchase_data["Gender"].value_counts()
purgen_df["Average Total Purchase Per Person"] = per_person

purgen_df.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 49.9]
age_range = ["10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels=age_range)

agegroup_count = purchase_data.groupby("Age Range")["SN"].nunique()
agegroup_per = (agegroup_count/total_player_count)*100

age_demo_df = pd.DataFrame({
    "Total Players": agegroup_count,
    "Percentage of Players": agegroup_per.round(2).map("{:.2f}%".format)        
})

age_demo_df                                             
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python


pur_age = purchase_data.groupby("Age Range")["SN"].count()

avg_pur_price = purchase_data.groupby("Age Range")["Price"].mean().map("${:.2f}".format)

total_pur_value = purchase_data.groupby("Age Range")["Price"].sum().map("${:.2f}".format)

avg_pur_per_person = (purchase_data.groupby("Age Range")["Price"].sum()/pur_age).map("${:.2f}".format)

pur_analysis_sum_df = pd.DataFrame({"Purchase Count": pur_age,
                                   "Average Purchase Price": avg_pur_price,
                                   "Total Purchase Value": total_pur_value,
                                   "Avg Total Purchase per Person": avg_pur_per_person})


pur_analysis_sum_df
```
